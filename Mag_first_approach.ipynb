```python
import json
import pandas as pd
import time
import glob
```

```python
path = r'/home/labnauk/prepared_data'
#all_files = glob.glob(path + "/mag_papers_0*.txt")
all_files = []
all_files.append(path + "/mag1000.txt")
```

```python
time2V1 = time.clock()

data=[]

for filename in all_files:
    with open(filename) as f:
        for line in f:
            data.append(json.loads(line))

time2V2 = time.clock()
```

```python
key_words = df_clean["keywords"]

key_words_list = []
for values in key_words:
    values = values.split(',')
    for value in values:
        key_words_list.append(value.lower())
```

```python
df_keyword = pd.DataFrame(key_words_list, columns=['key word'])
df_keyword['key word'].value_counts().to_csv(r'/home/labnauk/data_analysis/keyword_counter.csv')
ile_k = len(df_keyword['key word'].value_counts())
```

```python
from collections import Counter
import numpy as np

counter = Counter(df_keyword['key word'])
key_w = counter.keys()
key_w_c = counter.values()

indexes = np.arange(len(key_w))
width = 0.7
```

```python
df_n_citation = df_clean[df_clean['keywords'].str.contains(df_keyword["key word"].value_counts().index.tolist()[0])]
```

```python
n={}
for values in df_n_citation["year"]:
    n["{0}".format(values)] = df_n_citation.loc[df_n_citation["year"] == values]
```

```python
n_c = []
year_n_c = []    
for i in range(len(n)):
    n_c.append(sum(list(n.items())[i][1]["n_citation"].values.tolist()))
    year_n_c.append(list(n.items())[i][1]["year"].values.tolist()[0])
```

```python
common_v = []
common2_v = []
for i in range(len(d)):
    common_v.append(key_words_divided.iloc[i].mode()[0])
    if len(key_words_divided.iloc[i].mode()) >= 2:
        common2_v.append(key_words_divided.iloc[i].mode()[1])
    else:
        common2_v.append("")
```
